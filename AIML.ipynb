# Lab 1

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data ={'HoursStudied': [0, 1, 2, 3, 4, 5], 'Marks': [100, 40, 30, 20, 10, -20]}
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['HoursStudied'], df['Marks'])
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

```python
X = df[['HoursStudied']]
y = df['Marks']
model = LinearRegression()
model.fit(X, y)
```

```python
new_data = pd.DataFrame({'HoursStudied': [6]})
prediction = model.predict(new_data)
print("Predicted Marks:", prediction[0])
```

```python
import numpy as np
import pandas as pd

data = {
    "Age": [22 , 25 , 28 , 30 , np.nan , 35 , 40, 45, 50],
    "Marks": [81, 82, 90, 75, 92, 85.5, 91, 84, 200],
    "Gender": ["Male","Female", "Male", None, "Female", "Male", "Male", "Female", "Male"]
}

df = pd.DataFrame(data)

df["Passed"] = df["Marks"] >= 40


def assign_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "D"

df["Grade"] = df["Marks"].apply(assign_grade)

print(df)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
print("Mean of Age:", df["Age"].mean())
```

```python
df["Marks"] = df["Marks"].fillna(df["Marks"].median())
print("Median of Marks:", df["Marks"].median())
print(df["Gender"].mode()[0])
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
print(df)
```

```python
Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)
IQR = Q3 - Q1
print("IQR of Marks:", IQR)
lowb = Q1 - 1.5 * IQR
upb = Q3 + 1.5 * IQR
print("Lower Bound for Outliers:", lowb)
print("Upper Bound for Outliers:", upb)
outliers = df[(df["Marks"] < lowb) | (df["Marks"] > upb)]
print("Outliers:")
print(outliers)
```

```python
df[(df["Marks"] < lowb) | (df["Marks"] > upb)]
df_removed = df[(df["Marks"] >= lowb) & (df["Marks"] <= upb)]
print(df_removed)
```

```python
print("Original Size:", len(df))
print("After Removing:", len(df_removed))
```

```python
df_Capped = df.copy()
df_Capped["Marks"] = np.where(
    df_Capped["Marks"] > upb,
    upb,
    df_Capped["Marks"]
)
print(df_Capped)
```

```python
print("Original mean:", df["Marks"].mean())
print("After Removing:", df_removed["Marks"].mean())
print("After Capping Mean:", df_Capped["Marks"].mean())
```
